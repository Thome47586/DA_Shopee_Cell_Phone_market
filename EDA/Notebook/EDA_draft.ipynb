```python
# def function clean name của 2 bảng

def clean_name(series):
    series_name =   series.lower()

    if '\n' in series_name:
        series_name = series_name.split('\n')[-1]

    if "tr]" in series_name:
        series_name = series_name.split("tr]")[-1]

    if "k]" in series_name:
        series_name = series_name.split("k]")[-1]

    if series_name[0:1] == ' ':
        series_name = series_name[1:]

    pattern = r'điện thoại .+'
    check = re.findall(pattern, series_name)
    if len(check) == 0:
        return series_name
    else:
        return check[0]
```
